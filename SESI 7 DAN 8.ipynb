```python
#Mapping 100 row pertama dari tabel insiden
df_incidents_100 = df_incidents.iloc[0:100,:]
df_incidents_100
```

```python
df_incidents_100
```

```python
#San Fransisco Lat and Lon
latitude= 37.77 #garis lintang
longitude=-122.42 #garis bujur
```

```python
sf = folium.Map(location=[latitude,longitude],zoom_start=12)
sf
```

```python
sf = folium.Map(location=[latitude,longitude],zoom_start=12)
sf
incidents = folium.map.FeatureGroup()

# add 100 crimes to FeatureGroup

for lat,lon in zip(df_incidents_100['Y'],df_incidents_100['X']):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        
        )
    
    )
sf.add_child(incidents)
```

```python
df_incidents_100.Category
```

```python
# Clustering semua data berdasarkan kemiripan
from folium import plugins
sf = folium.Map(location=[latitude,longitude],zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sf)

for lat,lon,label in zip(df_incidents_100['Y'],df_incidents_100['X'],df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        
        )
    
    )
sf.add_child(incidents) 
```
