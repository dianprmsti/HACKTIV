# Normal Distribution/Gaussian Distribution
mempunyai bentuk grafik simetris dan berbentuk lonceng/kurva (pengukuran tinggi badan, berat badan)

parameter utama distribusi normal:

μ = Rata-Rata (menunjukkan pusat kurva normal )

σ = Standar Deviasi (menunjukkan seberapa tersebar datanya)

Semakin besar nilai standard deviation, semakin tersebar datanya, sementara semakin kecil nilai standard deviation, semakin dekat data dengan mean.

```python
from scipy.stats import norm

data_normal = norm.rvs(size=10000,loc=0,scale=1)

sns.distplot(data_normal)
```

# Gamma Distribution (Continuous)
distribusi probabilitas yang digunakan untuk menggambarkan waktu tunggu antara peristiwa yang terjadi dalam suatu proses acak, di mana peristiwa-peristiwa ini terjadi secara independen dan secara konstan dalam waktu. Gamma distribution didefinisikan oleh dua parameter:
1. alpha (α) (mengontrol bentuk dan skala kurva)

2. beta (β) (mengontrol posisi kurva pada sumbu x)

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(size=10000,a=100)
sns.distplot(data_gamma)
```

```python
len(gamma.rvs(size=10000,a=5))
```

# Exponential Distribution
distribusi probabilitas yang digunakan untuk menggambarkan waktu tunggu antara peristiwa yang terjadi secara independen dalam suatu proses acak, di mana waktu tunggu antara peristiwa-peristiwa ini memiliki pola eksponensial. didefinisikan oleh satu parameter, yaitu lambda (λ) yang mengontrol kecepatan dengan mana pola eksponensial memudar. Semakin besar nilai lambda, semakin cepat pola eksponensial memudar. 

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)
sns.distplot(data_expon)
```

# Poisson Distribution (Discrete/Count Data)
distribusi probabilitas diskret yang digunakan untuk menggambarkan frekuensi kejadian acak yang jarang terjadi dalam suatu periode waktu atau suatu wilayah. Contohnya bisa digunakan untuk menghitung jumlah kendaraan yang lewat di suatu jalan dalam satu jam, jumlah kerusakan mesin dalam satu hari, atau jumlah insiden kecelakaan dalam satu tahun. Poisson distribution didefinisikan oleh satu parameter, yaitu lambda (λ), yang merepresentasikan rata-rata frekuensi kejadian acak dalam periode waktu tertentu atau suatu wilayah. Semakin besar nilai lambda, semakin banyak frekuensi kejadian yang diharapkan terjadi.

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3,size=10000)
sns.distplot(data_poisson)
plt.xlabel('Poisson Distribution (Count Data)')
```

# Binomial Distribution (Discrete)
distribusi probabilitas diskret yang digunakan untuk menggambarkan hasil dari suatu percobaan yang terdiri dari n percobaan independen dengan hasil yang hanya bisa menghasilkan dua kemungkinan, yaitu sukses atau gagal.  Contohnya bisa digunakan untuk menghitung jumlah mahasiswa yang lulus ujian dalam kelas yang terdiri dari n mahasiswa, atau jumlah produk cacat dalam suatu pabrik. Didefiniskan 2 parameter yaitu:
1. n (merepresentasikan jumlah percobaan independen)
2. p (merepresentasikan probabilitas keberhasilan dalam satu percobaan)

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
sns.distplot(data_binom)
```

# Bernoulli Distribution (Discrete)
memiliki 2 kemungkinan yaitu 1(berhasil ) dan 0(gagal) dan dalam satu kali percobaan. Bernoulli distribution didefinisikan oleh satu parameter, yaitu p, yang merepresentasikan probabilitas keberhasilan dalam satu percobaan independen. Kurva Bernoulli distribution terdiri dari dua titik, yaitu p pada nilai sukses dan 1-p pada nilai gagal.

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)
sns.distplot(data_bern)
```

# Confidence Interval (CI)
rentang 2 nilai dimana nilai suatu sample mean tepat berada di tengah-tengahnya.

CI dinyatakan dalam batas atas dan batas bawah.
Contoh 160 +- 5

Batas Bawah 155

Batas Atas 165

Confidence Interval 95% Maka Z-Score 1.96

## Confidence Interval for the  Population Proportion

```python
import pandas as pd
import numpy as np

# membaca dataset
df = pd.read_csv('/content/drive/MyDrive/msib/msib/heart.csv')
df.head()
```

Sex 1=Pria dan 0=wanita

Target 1=Sakit Jantung , 0=bukan Penderita Sakit Jantung

```python
def sex_label(x):
    if x==1:
        return 'Male'
    else:
        return 'Female'
```

```python
df['sex_label'] = df['sex'].apply(sex_label)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.target.value_counts()
```

```python
df.groupby(['sex_label','target'])[['target']].count()
```

```python
# menghitung proporsi penduduk wanita yang menderita penyakit jantung
n = 86+226
pro_fm = 226/n
print(n,pro_fm)
```

```python
# standard error dihitung dengan membagi deviasi standar dari populasi dengan akar kuadrat dari ukuran sampel. 
# standard error penduduk wanita 
stderror_fm = (pro_fm*(1-pro_fm)/n)**(1/2)
stderror_fm
```

```python
# membuat CI dengan z-score 1.96 untuk confidence interval 95%
z_score = 1.96
low_ci_fm = pro_fm - (z_score*stderror_fm) #lower limit of CI
up_ci_fm = pro_fm + (z_score*stderror_fm) #upper limit of CI
print(low_ci_fm,up_ci_fm)
```

```python
#  Margin of error atau batas kesalahan adalah ukuran statistik yang digunakan untuk mengukur seberapa dekat hasil dari sampel acak dengan hasil yang sebenarnya dari populasi. Margin of error dapat dianggap sebagai selisih antara hasil estimasi sampel dan nilai sebenarnya dari populasi.
print('Margin Error=',(z_score*stderror_fm))
```

## Visualisasikan Area dari Confidence Interval

```python
stats.norm.cdf(2.17)
```

```python
z = 1.96
```

```python
lower_bound = mean_of_sample_means - z * standard_error
lower_bound
```

```python
kde = stats.gaussian_kde(sample_means)
pos = np.linspace(np.min(sample_means),np.max(sample_means),10000)
plt.plot(pos,kde(pos))
shade = np.linspace(lower_bound,upper_bound,300)
plt.fill_between(shade,kde(shade),alpha=0.5)
plt.xlabel('Sample Mass (Pounds)')
plt.ylabel('Frequency')
plt.axvline(x=mean_of_sample_means,color='red')
plt.text(x=152,y=.01,s='Shaded area are \npart of the \nconfidence intervals \n95%')
```

# Hypothesis Testing
suatu teknik statistik inferensial yang digunakan untuk memeriksa apakah suatu pernyataan tentang suatu populasi adalah benar atau salah, berdasarkan sampel acak dari populasi tersebut. Terdapat dua hipotesis: 
1. hipotesis nol/null hypothesis(menyatakan bahwa tidak ada perbedaan yang signifikan antara parameter populasi dan nilai tertentu yang diasumsikan)
2. hipotesis alternatif/alternative hypothesis (menyatakan bahwa ada perbedaan yang signifikan antara parameter populasi dan nilai tertentu yang diasumsikan)

alpha biasanya sama dengan 0.05 karena confidence level 95%

Apabila confidence level 99% maka alpha 0.01

# Hypothesis Testing in Population Means

```python
df = pd.read_csv('/content/drive/MyDrive/msib/msib/nhanes_2015_2016.csv')
df
```

Uji apakah rata2 BMI dari Laki-Laki lebih tinggi dari pada rata2 BMI Perempuan

h0 = rata2 BMI Laki-Laki tidak lebih tinggi signifikan dari perempuan

h1 = rata2 BMI Laki2 lebih tinggi signifikan dari perempuan

h0 : meanL <= meanP

h1: meanL > meanP

RIAGENDR female = 2 Male=1

```python
females = df[df.RIAGENDR==2]
males = df[df.RIAGENDR==1]
```

```python
# jumlah data, mean, stdeviasi
n_females = len(females)
mean_females = females['BMXBMI'].mean()
std_females = females['BMXBMI'].std()
print(n_females,mean_females,std_females)
```

```python
# jumlah data, mean, stdeviasi
n_males = len(males)
mean_males = males['BMXBMI'].mean()
std_males = males['BMXBMI'].std()
print(n_males,mean_males,std_males)
```

h0 : meanL <= meanP

h1: meanL > meanP

```python
import statsmodels.api as sm
```

```python
#karena confidence level 95% maka alpha=0.05
alpha = 0.05
```

```python
bmi_hypo = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='larger')
bmi_hypo
```

```python
if bmi_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0) Mean of Males BMI is not Larger than Females')
else:
    print('Reject Null Hypothesis (h0)  Mean of Males BMI is Larger than Females')
```

# Data sama hipotesis berbeda

Uji apakah mean BMI Laki Laki tidak sama dengan mean BMI Perempuan

h0 : meanL = meanP

h1: meanL != meanP

```python
bmi_hypo_2s = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='two-sided')
bmi_hypo_2s
```

```python
if bmi_hypo_2s[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```

# Hypothesis Testing Population Proportion


```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
```

```python
pro_hypo = sm.stats.proportions_ztest(n*pro_alt,n,pro_null,alternative='larger')
pro_hypo
```

```python
if pro_hypo[1]>alpha:
    print('Accept Null Hypothesis (h0)')
else:
    print('Reject Null Hypothesis (h0)')
```
