# Classification

Klasifikasi adalah proses mengelompokkan data ke dalam kategori yang sudah ditentukan berdasarkan ciri-ciri tertentu. Terdapat dua jenis algoritma klasifikasi, yaitu yang menggunakan data pelatihan berlabel (supervisi) dan yang tidak menggunakan data pelatihan berlabel (tidak supervisi).

# Logistic Regression 

1.PNG

metode dalam pembelajaran mesin yang digunakan untuk memprediksi kemungkinan terjadinya suatu peristiwa atau kejadian berdasarkan data input.

tipe:
1. binary logistic regression (variabel dependen hanya memiliki 2 kemungkinan hasil/kelas)
2. multinomial logistic regression (variabel dependen hanya memiliki 2 kemungkinan hasil/kelas 3 atau lebih tanpa urutan)
3. ordinal logistic regression (variabel dependen hanya memiliki 2 3 atau lebih kemungkinan hasil/kelas dengan pengurutan)

# Evaluation Metrics

Confusion matrix adalah tabel yang digunakan untuk mengevaluasi kinerja model klasifikasi dengan membandingkan nilai aktual dan nilai prediksi.

2.PNG


Empat kemungkinan jenis hasil dalam binary classification :
1. True Positive (TP): Data yang diklasifikasikan sebagai positif oleh model dan benar-benar positif dalam data aktual.
2. False Positive (FP): Data yang diklasifikasikan sebagai positif oleh model tetapi sebenarnya negatif dalam data aktual.
3. True Negative (TN): Data yang diklasifikasikan sebagai negatif oleh model dan benar-benar negatif dalam data aktual.
4. False Negative (FN): Data yang diklasifikasikan sebagai negatif oleh model tetapi sebenarnya positif dalam data aktual.

contoh:

kelas = 1 dan 0

1 itu spam

0 bukan spam

True Positive = Predicted Spam dan Aktual Spam

False Positive = Prediksi Spam dan Aktual bukan Spam

True Negative = Prediksi Bukan Spam dan Aktual Bukan Spam

False Negative = Prediksi Bukan Spam ternyata Aktual Spam

Metrics:

Akurasi = (True Positive + True Negative) / Jumlah Keseluruhan Data

## Logistic Regression for Handwriting Recognition

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report , confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x,y  = load_digits(return_X_y=True)
```

```python
print(x,'\n',y)
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=2024)
```

```python
x_train
```

6.PNG

```python
scaler = StandardScaler()
x_train  = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression(solver='liblinear',C=0.05,multi_class='ovr',random_state=2024)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_train,y_pred_train)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
cm = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
print('Train Set')
print(classification_report(y_train,y_pred_train))
print('Test Set')
print(classification_report(y_test,y_pred_test))
```

## K-Nearest Neighbour

8.PNG

```python
#Feature 1
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy',
          'Sunny','Overcast','Overcast','Rainy']
# Feature 2
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool',
       'Mild','Mild','Mild','Hot','Mild']

# Label atau target
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

print(len(weather),len(temp),len(play))
```

weather = Sunny:2, Rainy:1, Overcast: 0

temp = Mild :2 , Hot: 1 , Cool:0

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
weather_enc = le.fit_transform(weather) #var 1
temp_enc = le.fit_transform(temp) #var 2
play_enc = le.fit_transform(play) #var target
print(weather_enc)
print(temp_enc)
print(play_enc)
```

```python
features = list(zip(weather_enc,temp_enc))
features
```

```python
from sklearn.neighbors import KNeighborsClassifier
x = features
y  = play_enc
model = KNeighborsClassifier(n_neighbors=3) #n_neighbors adalah jumlah k

model.fit(x,y)
```

weather = Sunny:2, Rainy:1, Overcast: 0

temp = Mild :2 , Hot: 1 , Cool:0

play = Yes:1 , No:0

Ketika kondisi Overcast dan Temperature Mild akan Play atau tidak?

```python
predicted = model.predict([[1,1]])
print(predicted)
```

```python
y_pred = model.predict(x)
```

```python
print(classification_report(y,y_pred))
```

```python
import pandas as pd
```

```python
x = pd.DataFrame(wine['data'],columns=wine['feature_names'])
x
```

```python
y = pd.DataFrame(wine['target'],columns=['kultivar'])
y
```

```python
X_train , X_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=2024)
```

Coba K=5

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy :', accuracy_score(y_test,y_pred))
```

```python
knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_7.fit(X_train,y_train)

y_pred_7 = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy :', accuracy_score(y_test,y_pred_7))
```

```python
y_pred
```

```python
error = []

# kalkulasi eror untuk k_values between 1 dan 40

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i !=y_test['kultivar']))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,40),error, color='red',linestyle='dashed',marker='o', markerfacecolor='blue',markersize=12)
plt.title('Error rate K')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn_11 = KNeighborsClassifier(n_neighbors=11)
knn_11.fit(X_train,y_train)

y_pred_11 = knn_11.predict(X_test)

print('Accuracy:', accuracy_score(y_test,y_pred_11) )
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn_11 = KNeighborsClassifier(n_neighbors=11)
knn_11.fit(X_train_scaled,y_train)
y_pred_11_scaled = knn_11.predict(X_test_scaled)
print('Accuracy:', accuracy_score(y_test,y_pred_11_scaled) )
```

```python
y_pred_train = knn_11.predict(X_train_scaled)
y_pred_test = knn_11.predict(X_test_scaled)
```

```python
print('Train Set')
print(classification_report(y_train,y_pred_train))
print('Test Set')
print(classification_report(y_test,y_pred_test))
```
