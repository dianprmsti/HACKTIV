## Missing Value Category

### A.3.1. Missing Completely at Random (MCAR)

jenis missing data di mana probabilitas terjadinya nilai yang hilang sama untuk semua pengamatan atau baris dalam dataset. Dalam hal ini, hilangnya data tidak terkait dengan nilai lain dalam dataset dan tidak ada pola atau pengaruh tertentu yang menyebabkan hilangnya data.

Karakteristik dari Missing Completely At Random (MCAR) adalah:

- Probabilitas kehilangan nilai yang sama untuk semua observasi (baris).
- Tidak ada hubungan sama sekali antara data yang hilang dan nilai lain, baik yang teramati atau yang hilang, dalam dataset. Dengan kata lain, data yang hilang tersebut adalah subset acak dari data. Tidak ada yang sistematis yang membuat beberapa data lebih mungkin hilang daripada yang lain.
- Mengabaikan kasus-kasus tersebut tidak akan memengaruhi inferensi yang dibuat (hanya jika jumlah total nilai yang hilang kecil).

### Missing Not At Random (MNAR)

jenis missing data di mana kemungkinan terjadinya nilai yang hilang bergantung pada nilai yang hilang itu sendiri, yang tidak dapat dijelaskan oleh variabel yang tersedia. Dengan kata lain, dalam kasus MNAR, nilai yang hilang memiliki hubungan dengan variabel yang tidak terukur, yang membuatnya sulit untuk memprediksi nilai yang hilang.

Karakteristik dari Missing Not At Random (MNAR) adalah:

- Ada mekanisme atau alasan mengapa nilai yang hilang muncul dalam dataset.
- Kehilangan nilai tersebut tergantung pada informasi yang tidak tercatat dalam dataset.
- Dalam situasi ini, kita lebih baik menandai nilai yang hilang untuk memprediksi target kita.

## Missing Value Handling Methods

### Complete Case Analysis

Complete-case analysis (CCA), juga disebut penghapusan kasus dengan cara daftar, terdiri dari membuang observasi di mana nilai pada salah satu variabel hilang. Analisis Kasus Lengkap secara harfiah berarti menganalisis hanya observasi untuk yang terdapat informasi di semua variabel (Xs).

CCA dapat diterapkan pada variabel kategorikal dan numerik.

Asumsi :
- CCA berfungsi dengan baik ketika data hilang sepenuhnya secara acak (MCAR).
- Dalam hal ini, mengeluarkan observasi dengan data hilang sama dengan secara acak mengeluarkan beberapa observasi dari dataset, mengingat bahwa data hilang benar-benar acak.
- Berfungsi dengan baik jika total jumlah nilai yang hilang kecil.

Keuntungan :
- Mudah untuk diimplementasikan
- Set data yang sama (meskipun set data yang lebih kecil) digunakan untuk semua analisis (tidak ada manipulasi data)
- Menjaga distribusi variabel (jika data MCAR, maka distribusi variabel pada dataset yang lebih kecil harus sesuai dengan distribusi pada dataset asli)

Kerugian :
- Ini dapat mengeluarkan sebagian besar sampel asli, yang sebenarnya informatif untuk analisis
- CCA akan bias jika kasus lengkap secara sistematis berbeda dari sampel asli (misalnya ketika informasi yang hilang sebenarnya adalah MAR (Missing At Random) atau MNAR (Missing Not At Random)).

### Mean Median Imputation

**Imputasi adalah tindakan mengganti data yang hilang dengan perkiraan statistik dari nilai yang hilang.** Tujuan dari setiap teknik imputasi adalah menghasilkan dataset lengkap yang dapat digunakan untuk machine learning atau deep learning.

Imputasi mean/median terdiri dari mengganti semua kemunculan nilai yang hilang (NA) dalam suatu variabel dengan:
* **Mean** : jika variabel memiliki **distribusi Normal/Gaussian**.
* **Median** : jika variabel memiliki **distribusi condong/skewed**.

**Asumsi** : Data hilang sepenuhnya secara acak (MCAR).

Jika demikian, kita dapat mengganti NA dengan kemunculan yang paling sering dari variabel, yang merupakan mean jika variabel memiliki distribusi Gaussian, atau median jika tidak.

**Keuntungan** :
* Mudah untuk diimplementasikan
* Cara yang cepat untuk mendapatkan dataset lengkap

**Keterbatasan** :
* Distorsi varians asli

  Saat mengganti NA dengan mean atau median, varians variabel akan terdistorsi jika jumlah NA besar dibandingkan dengan total jumlah observasi (karena nilai yang diimputasi tidak berbeda dari mean atau satu sama lain). Oleh karena itu, dapat menyebabkan penghentian estimasi varians.

* Distorsi kovarians/korelasi dengan variabel lain dalam dataset

  Selain itu, estimasi kovarians dan korelasi dengan variabel lain dalam dataset juga dapat terpengaruh. Hal ini terjadi karena kita dapat menghancurkan korelasi intrinsik karena mean/median yang sekarang menggantikan NA tidak akan mempertahankan hubungan dengan variabel lain.

**Imputasi harus dilakukan pada set pelatihan, dan kemudian dipropagasikan ke set pengujian.** Ini berarti bahwa mean/median yang digunakan untuk mengisi nilai yang hilang baik dalam set pelatihan maupun pengujian, harus diekstraksi dari set pelatihan saja.

## Random Sample Imputation

Random sampling terdiri dari mengambil observasi acak dari kolam observasi yang tersedia dari variabel tersebut, dan menggunakan nilai yang diekstraksi secara acak untuk mengisi nilai yang hilang.

Imputasi Sampel Acak pada dasarnya mirip dengan imputasi rata-rata/median, dalam arti bahwa tujuannya adalah untuk mempertahankan parameter statistik dari variabel asli, di mana data hilang.

Dengan menyampling secara acak observasi variabel untuk instance di mana data tersedia, kami menjamin bahwa rata-rata dan standar deviasi variabel dipertahankan.

Asumsi:

- Data hilang sepenuhnya acak (MCAR).

  Jika demikian, masuk akal untuk mengganti nilai yang hilang, dengan nilai yang diekstraksi dari distribusi variabel asli.

- Dari sudut pandang probabilitas, nilai yang lebih sering muncul (seperti mean atau median) akan dipilih lebih sering (karena ada lebih banyak nilai yang dapat dipilih), tetapi nilai lain yang lebih jarang juga akan dipilih. Dengan demikian, variansi variabel tetap dipertahankan. Rasionale-nya adalah untuk menggantikan populasi nilai yang hilang dengan populasi nilai dengan distribusi yang sama dengan variabel.

Keuntungan:

- Mudah untuk diimplementasikan
- Cara cepat untuk mendapatkan dataset yang lengkap
- Memperoleh variasi variabel yang dipertahankan

Keterbatasan : Keacakan

Keacakan mungkin tidak terlalu menjadi perhatian saat mengganti nilai yang hilang untuk kompetisi data, di mana seluruh batch nilai yang hilang diganti sekali, kemudian dataset dinilai dan itu adalah akhir dari masalah. Namun, dalam skenario bisnis, situasinya sangat berbeda.

```python
def impute_na_rand(df,variable,mean_value,median_value):
    df[variable + '_mean'] = df[variable].fillna(mean_value)
    df[variable + '_median'] = df[variable].fillna(median_value)
    df[variable + '_zero'] = df[variable].fillna(0)
    df[variable + '_random'] = df[variable]
    
    ## Extract Random Sample to impute Na
    random_sample = df[variable].dropna().sample(df[variable].isna().sum(),random_state=0)
    
    ## Pandas needs to have same index in order merge dataset
    random_sample.index = df[df[variable].isna()].index
    
    df.loc[df[variable].isnull(),variable+'_random']=random_sample
    
    return df
    
```

```python
X_train = impute_na_rand(X_train,'Age',age_mean,age_median)
X_test = impute_na_rand(X_test,'Age',age_mean,age_median)
X_train
```

### Adding Variable to Capture Missing Values

Pada kuliah sebelumnya, kita mempelajari cara mengganti nilai yang hilang dengan imputasi mean/median atau dengan mengambil contoh acak dari variabel untuk instance yang memiliki data yang tersedia, dan menggunakan nilai tersebut untuk menggantikan nilai yang hilang. Kita juga membahas bahwa kedua metode ini mengasumsikan bahwa data yang hilang hilang secara lengkap acak (MCAR). Lalu bagaimana jika data tidak hilang secara lengkap acak? Dengan menggunakan prosedur ini, kita akan kehilangan informasi prediktif yang penting.

Bagaimana cara mencegah hal tersebut?

Kita dapat menangkap pentingnya hilangnya data dengan membuat variabel tambahan yang menunjukkan apakah data hilang untuk pengamatan tersebut (1) atau tidak (0). Variabel tambahan ini merupakan variabel biner:
- 0 menunjukkan bahwa nilai hadir untuk pengamatan itu, dan
- 1 menunjukkan bahwa nilai hilang untuk pengamatan itu.

Keuntungan:

# Arbitrary Imputation

### Explanation

In case of missing values are not MCAR, we can use several methods below:

* **Adding an additional binary variable** to indicate whether the value is missing (`1`) or not (`0`).

* **Replacing the NA by a value** 
  - Sometimes this value at a far end of the distribution
  - This method is used in several Kaggle competitions. 
  - It consists of replacing the NA by an arbitrary value. Any of your creation, but ideally different from the median/mean/mode.

**Advantages :**
* Easy to implement.
* Captures the importance of missingess if there is one.

**Disadvantages :**
* Distorts the original distribution of the variable.
* If missingess is not important, it may mask the predictive power of the original variable by distorting its distribution.
* Hard to decide which value to use.
 If the value is outside the distribution it may mask or create outliers

When variables are captured by third parties, like credit agencies, they place arbitrary numbers already to signal this missingness. So,sometimes this technique is **common practice in real life data collections**.

## Guidelines

* Check of missing data mechanism :
  - MCAR : 
    * Complete Case Analysis, 
    * Mean Imputaion (numeric variable with Gaussian/Normal Distribution), 
    * Median Impuation (numeric variable with skewed distribution)
    * Mode Imputation (categorical variable)
    * Add Missing Label Imputation (categorical variable)
    * Random Sample Imputation
    * Arbitray Value Imputation
  - MAR, MNAR : 
    * Add 0/1 Imputation 
    * Arbitray Value Imputation
    * Add Missing Label Imputation (categorical variable)
    * You can combined it with Mean/Median/Mode Imputation
* If missing values are too small : Complete Case Analysis
* If missing values are less than 5% of the variable :
  - Mean Imputation
  - Median Imputation
  - Mode Imputation
  - Random Sample Imputation
* If missing values are more than 5% of the variable:
  - Mean/Median/Mode Imputation + adding 0/1 Imputation
  - Add Missing Label Imputation (categorical variable)

If the number of NA in a variable is small, they are unlikely to have a strong impact on the variable / target that you are trying to predict. Therefore, treating them specially, will most certainly add noise to the variables. Therefore, it is more useful to replace by mean/random sample to preserve the variable distribution.

If the variable / target you are trying to predict is however highly unbalanced, then it might be the case that this small number of NA are indeed informative. You would have to check this out.

---
## Final Note

NA imputation for data competitions and business settings can be approached differently. **In data competitions, a tiny increase in performance can be the difference between 1st or 2nd place.** Therefore, you may want to **try all the feature engineering methods** and use the one that gives the best machine learning model performance. It may be the case that different NA imputation methods help different models make better predictions.

In business scenarios, scientist **don't usually have the time to do lengthy studies**, and may therefore choose to streamline the feature engineering procedure. In these cases, it is common practice to follow the guidelines above, taking into account the exceptions, and do the same processing for all features.

This streamlined pre-processing may not lead to the most predictive features possible, yet it makes feature engineering and machine learning models delivery substantially faster. Thus, the business can start enjoying the power of machine learning sooner.

```python
# Buat function untuk kolom-kolom yang terdistribusi normal dan skewed

#Normal
def find_normal_boundaries(df,variable):
    upper_boundary = df[variable].mean() + 3 * df[variable].std()
    lower_boundary = df[variable].mean() - 3 *df[variable].std()
    
    return upper_boundary,lower_boundary

# Skewed
def find_skewed_boundaries(df,variable):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)
    
    lower_boundary = df[variable].quantile(0.25) - (1.5 * IQR)
    upper_boundary = df[variable].quantile(0.75) +(1.5 * IQR)
    return upper_boundary,lower_boundary
```

# Trimming

## Trimming or truncation

Trimming, also known as truncation, involves removing the outliers from the dataset. We only need to decide on a metric to determine outliers. As we saw in section 3, this can be the Gaussian approximation for normally distributed variables or the inter-quantile range proximity rule for skewed variables.

### Advantages

- quick

### Limitations

- outliers for one variable could contain useful information in the other variables
- if there are outliers across many variables, we could remove a big chunk of dataset

## Important

Outliers should be detected AND **removed ONLY** from the training set, and NOT from the test set. So we should first divide our data set into train and tests, and remove outliers in the train set, but keep those in the test set, and measure how well our model is doing.

# Censoring or Capping

## Censoring or Capping.

**Censoring**, or **capping**, means capping the maximum and /or minimum of a distribution at an arbitrary value. On other words, values bigger or smaller than the arbitrarily determined ones are **censored**.

Capping can be done at both tails, or just one of the tails, depending on the variable and the user.

### Advantages

- does not remove data

### Limitations

- distorts the distributions of the variables
- distorts the relationships among variables

## Important

When doing capping, we tend to cap values both in train and test set. It is important to remember that the capping values MUST be derived from the train set. And then use those same values to cap the variables in the test set
